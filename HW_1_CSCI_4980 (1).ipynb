```python
url = "http://motioncore-umh.cs.umn.edu/4980/college.csv"
data = urllib.request.urlopen(url)
csv_bytes = data.read()
csv_string = csv_bytes.decode("utf8")
csv_data = io.StringIO(csv_string)
df = pd.read_csv(csv_data)




df
```

```python
df['Selective'] = np.where(df['Top10perc'] > 50, "Selective", "Not Selective")
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

```

```python
X = df
y = df['Selective'].to_numpy()


# Split the data

X = X.drop(['School', 'Private','Top10perc', 'Top25perc', 'Selective'], axis=1).to_numpy()


def train_test_split(X,y,test_size):
  numRows = y.size                     #number or rows in the entire dataset
  splitPoint = int((1-test_size)*numRows)  #the row index test_size% of the way through
  p = np.random.permutation(numRows)   #array for shuffling data
  X = X[p]
  y = y[p]

  #Training data array
  X_train = X[:splitPoint]  # training features
  y_train = y[:splitPoint]  # training labels

  #Testing data array
  X_test = X[splitPoint:]  # testing features
  y_test = y[splitPoint:] # testing labels

  return X_train, X_test, y_train, y_test



```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,0.33)
```

```python
def accuracy(classify, X_train, y_train, X_test, y_test):
  pred = classify(X_train, y_train, X_test)
  correct_pred = (pred == y_test)
  total_preds = y_test.size
  if total_preds == 0:
    return 0
  return correct_pred.sum()/total_preds
```

```python
def recall(classify, X_train, y_train, X_test, y_test):
    pred = classify(X_train, y_train, X_test)
    true_positives = (pred[y_test == "Selective"] == "Selective").sum()
    false_negatives = (pred[y_test == "Selective"] == "Not Selective").sum()
    if true_positives + false_negatives == 0:
        return 0
    return true_positives / (true_positives + false_negatives)
```

```python
def precision(classify, X_train, y_train, X_test, y_test):
    pred = classify(X_train, y_train, X_test)
    true_positives = (pred[y_test == "Selective"] == "Selective").sum()
    false_positives = (pred[y_test == "Not Selective"] == "Selective").sum()
    if true_positives + false_positives == 0:
        return 0
    return true_positives / (true_positives + false_positives)
```

```python
def f1score(classify, X_train, y_train, X_test, y_test):
  myrec = recall(classify, X_train, y_train, X_test, y_test)
  mypre = precision(classify, X_train, y_train, X_test, y_test)
  if myrec == 0 or mypre == 0:
    return 0
  overrec = 1/recall(classify, X_train, y_train, X_test, y_test)
  overpre = 1/precision(classify, X_train, y_train, X_test, y_test)
  score = 2/(overrec+overpre)
  return score

```

```python
def avg_accuracy(classify, X, y, count = 200, split=0.33):
  acc = 0
  for _ in range(0,count):
    X_train, X_test, y_train, y_test = train_test_split(X,y,split)
    acc += accuracy(classify, X_train, y_train, X_test, y_test)
  return acc/count
```

```python
def avg_recall(classify, X, y, count = 200, split=0.33):
  rec = 0
  for _ in range(0,count):
    X_train, X_test, y_train, y_test = train_test_split(X,y,split)
    rec += recall(classify, X_train, y_train, X_test, y_test)
  return rec/count
```

```python
def avg_precision(classify, X, y, count = 200, split=0.33):
  pre = 0
  for _ in range(0,count):
    X_train, X_test, y_train, y_test = train_test_split(X,y,split)
    pre += precision(classify, X_train, y_train, X_test, y_test)
  return pre/count
```

```python
def avg_f1(classify, X, y, count = 200, split=0.33):
  f1 = 0
  for _ in range(0,count):
    X_train, X_test, y_train, y_test = train_test_split(X,y,split)
    f1 += f1score(f1, X_train, y_train, X_test, y_test)
  return f1/count
```

```python
def knn_classifier1(X_train, y_train, X_test,k=1): #Q4 1) KNN classifier with small k
  # instantiate learning model and fit data
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)

  # predict the response
  pred = knn.predict(X_test)
  return pred
```

```python

knn_classifier1(X_train, y_train, X_test)
print("accuracy on testing data")
print(accuracy(knn_classifier1, X_train, y_train, X_test, y_test))
print("accuracy on training data")
print(accuracy(knn_classifier1, X_train, y_train, X_train, y_train))
print("recall on the testing data")
print(recall(knn_classifier1, X_train, y_train, X_test, y_test))
print("recall on the training data")
print(recall(knn_classifier1, X_train, y_train, X_train, y_train))
print("precision on the testing data")
print(precision(knn_classifier1, X_train, y_train, X_test, y_test))
print("precision on the training data")
print(precision(knn_classifier1, X_train, y_train, X_train, y_train))
print("f1 on testing data")
print(f1score(knn_classifier1, X_train, y_train, X_test, y_test))
print("f1 on training data")
print(f1score(knn_classifier1, X_train, y_train, X_train, y_train))

print(avg_accuracy(knn_classifier1, X, y))
print(avg_precision(knn_classifier1, X, y))
print(avg_recall(knn_classifier1, X, y))
```
